Let's plot the generated "mask of masks" used in the PCM experiments.

The randomization scheme to pick patterns outside coarse_J can be either "uniform" or "multilevel". If "multilevel", we should see a difference in each level.

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

def plot_mask(experiment_name, filename):
    masks_dir = Path(f"../../pcm_demo_output") / experiment_name / "masks"
    filepath = masks_dir / filename
    np.load(filepath)
    plt.imshow(np.load(filepath), cmap='gray')
    plt.show()
```

```python
plot_mask(
    "20260114_080321_silicon_512x512_multilevel_pnp_512_image_eta_1_not_good_not_finished",
    "sample_80.0_percent_coarse_J=7_multilevel_random.npy"
)
```

```python
plot_mask(
    "20260114_080321_silicon_512x512_multilevel_pnp_512_image_eta_1_not_good_not_finished",
    "sample_80.0_percent_coarse_J=8_multilevel_random.npy"
)
```

```python
plot_mask(
    "20260114_084354_silicon_512x512_multilevel_pnp_eta_10_example_ok",
    "sample_20.0_percent_coarse_J=7_multilevel_random.npy"
)
```
